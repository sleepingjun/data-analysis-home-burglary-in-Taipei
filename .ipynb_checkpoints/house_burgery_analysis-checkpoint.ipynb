I wanna have cols denoted year, month, and day, so I need to split occurrence date into three cols.

```python
def split_date(data,date_col):
    """
    to split date into year, month, day list
    :param data: dataframe
    :param date_col: str
    :return: list
    """
    year=[]
    month=[]
    day=[]
    for i in data[date_col]:
        year.append(str(i)[:-4])
        month.append(str(i)[-4:-2])
        day.append(str(i)[-2:])
        #print(f'year: {year} month: {month} day: {day}')
    return year,month,day
```

```python
year,month,day=split_date(data,"發生日期")
```

```python
data.insert(5,column="year",value=year)
data.insert(6,column="month",value=month)
data.insert(7,column="day",value=day)
```

First, I build a dict function to help me collect how many items I need are in the col.  
I know there is a package in python called Collection. It's convinient and useful. However, it's slow than building a dict myself. Hence, I prefer to collect by dict.

```python
def build_dict(col):
    """
    :param col: list
    :return: dict
    to build dict to count list
    """
    d=dict()
    for i in col:
        if i in d:
            d[i]+=1
        else:
            d[i]=1

    return dict((key,d[key]) for key in sorted(d.keys()))
```

I want to plot a pie with percentage.  

```python
def num_to_percent(num,gross=total):
    """
    number invert to percentage
    :param num: part num
    :param gross: total num
    :return: float
    """
    absolute = int(np.round(num/100.*(gross)))
    return absolute
```

```python
house_burglary_per=[]
for i in house_burglary_amount:
    house_burglary_per.append(num_to_percent(i))
```

```python
fig, ax =plt.subplots(figsize=(6,3))
wedges, texts, autotexts = ax.pie(house_burglary_per,
                                  labels=area_name,
                                  autopct='%1.1f%%'
                                  )

ax.legend(wedges, area_name,
          title="Area",
          loc="right",
          bbox_to_anchor=(1, 0, 0.5, 1),
          prop={'size':9})

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Administrative Area Percentage of House Burglary\nTotal: "+str(len(data)))
```
